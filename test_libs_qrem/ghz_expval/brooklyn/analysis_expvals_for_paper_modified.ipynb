# Expvals

```python
def expval(hist):
    if not hist:
        return 0
    expval = 0
    shots = 0
    for key, count in hist.items():
        shots += count
        sigma_z = 1
        for s in key:
            if s == "1":
                sigma_z *= -1
        expval += sigma_z * count
    expval /= shots
    return expval
```

```python
def plot_expvals(max_size,
                 range_list,
                 corrs_list,
                 stddevs_list,
                 labels,
                 title = None,
                 xlabel = "number of qubits",
                 ylabel = "expectation value"):
    plt.clf()
    plt.figure(dpi=200)
    plt.style.use('ggplot')
    linestyle = "solid"
    for i in range(len(corrs_list)):
        if i >= 7:
            linestyle = "dashed"
        if corrs_list[i] is None:
            continue
        corrs = np.asarray(corrs_list[i])
        if stddevs_list[i] is not None:
            stddevs = np.asarray(stddevs_list[i])
            plt.fill_between(range_list[i], corrs - stddevs, corrs + stddevs, alpha=0.2, color="C"+str(i + 1))
        if i < 2: #  or i == 4:
            plt.plot(range_list[i], corrs, label = labels[i], color="C"+str(i + 1), linestyle=linestyle, lw=3)
        elif i == 3:
            plt.plot(range_list[i], corrs, label = labels[i], color="C"+str(i + 1), linestyle="dashed")
        elif i == 5:
            plt.plot(range_list[i], corrs, label = labels[i], color="C"+str(i + 1), linestyle="dashed")
        else:
            plt.plot(range_list[i], corrs, label = labels[i], color="C"+str(i + 1), linestyle=linestyle)
    plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0.1, framealpha= 0.2, fontsize=14)
    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
    plt.ylim(-0.2, 1.2)
    plt.xticks(list(range(max_size + 2))[::5])
    plt.title(title)
    plt.show()
```
