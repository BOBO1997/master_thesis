```python
import numpy as np
import cmath
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
import importlib
```

```python
def P(n, shots, hist):
    # assert n == len(list(hist.keys())[0])
    count = 0
    count += hist.get("0" * n, 0) / shots
    count += hist.get("1" * n, 0) / shots
    return count
```

```python
def S_phi(n, shots, hist):
    # assert n == len(list(hist.keys())[0])
    return hist.get("0" * n, 0) / shots
```

```python
def C(n, shots, hists):
    # assert n == len(list(hists[0].keys())[0])
    iq = 0
    for i, hist in enumerate(hists):
        phi = (np.pi * i) / (n + 1)
        iq += cmath.exp(phi * n * 1j) * S_phi(n, shots, hist)
    iq = abs(iq) / len(hists)
    return 2 * np.sqrt(iq)
```

```python
max_size = 65
max_size = 39
shots = 8192
num_try = 8
print("max_size:", max_size)
print("shots:", shots)
print("num_try:", num_try)
```

```python
mooney001_fidelities_list = []
for n in range(1, max_size + 1):
    with open("./pkls/mooney001/info_list_" + str(n) + "-qubit.pkl", "rb") as f:
        mooney001_info_list = pickle.load(f)
    ps, cs = [], []
    fidelities = []
    for j in range(num_try):
        p = P(n, shots, mooney001_info_list[j]["ghz"]["mitigated_hist"])
        c = C(n, shots, [info["mitigated_hist"] for info in mooney001_info_list[j]["mqc"]])
        ps.append( p )
        cs.append( c )
        fidelities.append( (p + c) / 2 )
    mooney001_fidelities_list.append({"ps": ps, "cs": cs, "fidelities": fidelities })
    print(n, "qubit finished (average fidelity: ", np.mean(fidelities), ")")
```

```python
mooney001_S_phis_list_list = []
for n in range(1, max_size + 1):
    with open("./pkls/mooney001/info_list_" + str(n) + "-qubit.pkl", "rb") as f:
        mooney001_info_list = pickle.load(f)
    mooney001_S_phis_list = []
    for j in range(num_try):
        mooney001_S_phis = []
        for i, hist in enumerate([info["mitigated_hist"] for info in mooney001_info_list[j]["mqc"]]):
            mooney001_S_phis.append(S_phi(n, shots, hist))
        mooney001_S_phis_list.append(mooney001_S_phis)
    mooney001_S_phis_list_list.append(mooney001_S_phis_list)
    print(n, "qubit finished")
```

```python
with open("./pkls/mooney001/fidelity_info.pkl", "wb") as f:
    pickle.dump({"fidelities_list": mooney001_fidelities_list, "S_phis_list_list": mooney001_S_phis_list_list}, f)
```
